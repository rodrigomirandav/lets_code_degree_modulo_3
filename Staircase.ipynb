Explanation

The staircase is right-aligned, composed of # symbols and spaces, and has a height and width of .

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'staircase' function below.
#
# The function accepts INTEGER n as parameter.
#

def staircase(n):
    stair = ''

    for i in range(1, n):
        stair += ' ' * (n - i) + '#' * i + '\n'

    stair += "#" * n
    
    print(stair)

if __name__ == '__main__':
    n = int(input().strip())

    staircase(n)
```
